```python
ratings = pd.read_csv(
    'BX-Book-Ratings.csv',
    sep=';',
    encoding='latin-1',
    engine='python',
    on_bad_lines='skip'
)
```

```python
ratings.head()
```

```python
ratings.rename(columns={'User-ID':'user_id','ISBN':'isbn','Book-Rating':'rating'} , inplace=True)
```

```python
ratings.columns
```

```python
ratings['user_id'].value_counts().sum
```

```python
# Count number of ratings each user has
user_counts = ratings.groupby("user_id")["rating"].count()

# Filter: users with > 200 ratings
active_users = user_counts[user_counts > 200].index

# Filter original dataframe
ratings = ratings[ratings["user_id"].isin(active_users)]
```

```python
ratings.shape
```

```python
# Now join ratings with books
ratings_with_books = ratings.merge(books , on='isbn')
```

```python
ratings_with_books.head()
```

```python
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.head()
```

```python
ratings_with_books.shape
```

```python
final_rating = ratings_with_books.merge(number_rating , on='title')
```

```python
final_rating.head()
```

```python
final_rating.rename(columns={'rating_x':'rating','rating_y':'num_of_rating'}, inplace=True)
```

```python
final_rating.columns
```

```python
final_rating = final_rating[final_rating['num_of_rating']>=50]
```

```python
final_rating.head()
```

```python
final_rating.isnull().sum()
```

```python
# Let's create a pivot table
book_pivot = final_rating.pivot_table(columns='user_id' , index='title' , values='rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0 , inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
model.fit(book_sparse)
```

```python
distance , suggestion = model.kneighbors(book_pivot.iloc[237 , :].values.reshape(1 , -1) , n_neighbors=6)
```

```python
distance , suggestion
```
